```python
avg_price= df1.groupby("neighbourhood_group")["price"].mean().reset_index()
```

```python
avg_price
```

```python
plt.figure(figure=(10,6))
sns.barplot(data=avg_price, x = "neighbourhood_group", y= "price", palette='viridis')
plt.title("average price in Neighbourhood")
plt.xlabel("neighbourhood_group")
plt.ylabel("price")
plt.show()
```

```python
# As per the Bar chart we can notice that the prices of the AirBnb properties in Manhttan are the costliest followed by Queens and Brooklyn, 
# while least prices are seen in Bronx.
```

```python
unique_rooms = df1["room_type"].unique()
unique_rooms
```

```python
room_counts = df1["room_type"].value_counts()
room_counts
```

```python
plt.figure(figsize=(6,6))
colors = ("green", "orange", "yellow", "red")
plt.pie(room_counts, labels=room_counts.index, autopct='%1.1f%%', colors=colors)
plt.title("types of properties")
plt.show()
```

```python
# The aboce Pie chart shows that most of the hosts/owners own the entire home/apt which is 53.9%, 
# the second priority by owners/hosts is given to theprivate room,
# Shared room are used least.
```

```python
room_price= df1.groupby("room_type")["price"].mean().reset_index()
room_price
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(data= room_price, x ="room_type", y ="price")
plt.title("Average Prices Per Rooms", fontsize= 16, fontweight= "bold")
plt.xlabel("Room Types", fontsize= 12, fontweight= "bold")
plt.ylabel("Prices", fontsize= 12, fontweight= "bold")
plt.show()
```

```python
# By this Violine Chart we can easily see that the getting an entire room is the costliest, and shared rooms are the cheapest.
```

```python
reviews= df1.groupby("neighbourhood")["number_of_reviews"].mean().reset_index()
reviews
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data= reviews, x= "neighbourhood", y= "number_of_reviews")
plt.title("Outlighieng Neighbourhoods", fontsize=16, fontweight = "bold")
plt.xlabel("Neighbourhood", fontsize=12, fontweight = "bold")
plt.ylabel("Outlieghing Reviews", fontsize=12, fontweight = "bold")
plt.xticks(rotation= 90)
plt.show()
```

```python
# As per the chart the observation is that there are only two outlieghing review giving Neighbourhoods are there,
# they are East Flatbush and Long Island city.
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(data=df1, x="room_type", y="minimum_nights", color= "green")
plt.title("Nights Per Room Tyoe", fontsize=16, fontweight= "bold")
plt.xlabel("Type of room", fontsize=12, fontweight= "bold")
plt.ylabel("Night Stay", fontsize=12, fontweight= "bold")
plt.show()
```

```python
# We can observe that in entire home you can stay longer and for the short durations as well.
# In private rooms short and medium terms of stay plan.
# While in shared rooms you can stay not more then around 5 days.
```

```python
df2.drop(["MaritalStatusID", "FromDiversityJobFairID", "MarriedID", "GenderID"], axis=1, inplace=True)
df2
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Department', y='Salary', data=df2, estimator=np.mean)
plt.title('Average Salary by Department')
plt.xticks(rotation=45)
plt.ylabel('Average Salary')
plt.xlabel('Department')
plt.show()
```

```python
# Here we can observe that executive officers are earning the most and least is earned by production deparment.
# However the earning rate doesn't have much difference among all the departments except for Executive officers.
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='TermReason', hue='Department', data=df2)
plt.title('Termination Reasons by Department')
plt.ylabel('Count of Terminations')
plt.xlabel('Termination Reason')
plt.legend(title='Department')
plt.show()
```

```python
# most of the employees left the company as they got another position. 
# And the second major reason was because they were not happy. 
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='ManagerName', y='PerformanceScore', data=df2)
plt.title('Performance Score Distribution by Manager')
plt.xlabel('Manager Name')
plt.ylabel('Performance Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Most of the managers fully meets the expectation.
# There are three managers who might be not abe to perform very well.
# And there are some who needs to improve.
```

```python
absences_pivot = df2.pivot_table(values='Absences', index='Department', aggfunc='sum')

plt.figure(figsize=(10, 6))
sns.heatmap(absences_pivot, annot=True, cmap='YlGnBu', cbar=True, fmt='g')
plt.title('Heatmap of Total Absences by Department')
plt.xlabel('Department ID')
plt.ylabel('Total Absences')
plt.show()
```

```python
# By this we can observe that the employees of production  department are absent oftenly. 
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Salary', y='DaysLateLast30', data=df2, hue='RecruitmentSource', palette='viridis', s=100)
plt.title('Salary vs. Days Late')
plt.xlabel('Salary')
plt.ylabel('Days Late in Last 30 Days')
plt.legend(title='Recruitment Source', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.show()
```
